#####  Basic Operations on Images



```python
import cv2
import numpy as np
img = cv2.imread('mountain.jpg')
```

```python
cv2.namedWindow('mountain', cv2.WINDOW_NORMAL)
cv2.imshow('mountain',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
px = img[100,100]
print (px)
```

```python
# accessing only blue pixel
blue = img[100,100,0]
print (blue)
```

```python
img[100,100] = [255,255,255]
print (img[100,100])
```

```python
img[100:200,100:200] = [0,255,0]
print (img[100,100])
```

```python
cv2.namedWindow('mountain', cv2.WINDOW_NORMAL)
cv2.imshow('mountain',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img[500:700,200:400,2] = 255
```

```python
cv2.namedWindow('mountain', cv2.WINDOW_NORMAL)
cv2.imshow('mountain',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# accessing RED value
img.item(10,10,2)
```

```python
# modifying RED value
img.itemset((10,10,2),100)
img.item(10,10,2)
```
