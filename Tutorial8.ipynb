#####  Functions in Python 


```python
 def myfun(x, y):
    return x * y
myfun(3,4)
```

```python
def myfun(x):
    return x*3
def apply(q, x):
    return q(x)
apply(myfun, 7)
#함수도 argument 가능!
```

##### Lambda Notation


```python
apply(lambda z: z * 4, 7)
#함수가 들어가야하는데 lambda
#이름을 주고 함수형태로 해야하는데 한줄 짜리 함수일때는 이름을 안주고 그냥 lambda
#z가 argument고 z 곱하기 4
```
