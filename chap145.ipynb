# 1.4.5 Morphological Transformations

## Erosin

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('charj.png',0)
kernel = np.ones((5,5),np.uint8)
erosion1 = cv2.erode(img,kernel,iterations = 1)
erosion2 = cv2.erode(img,kernel,iterations = 2)
erosion3 = cv2.erode(img,kernel,iterations = 3)
```

```python
cv2.imshow('img',img)
cv2.imshow('dat1',erosion1)
cv2.imshow('dat2',erosion2)
cv2.imshow('dat3',erosion3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
